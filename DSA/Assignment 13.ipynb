### Answer 1

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def createNewLinkedList(list1, list2):
    if list1 is None or list2 is None:
        return None

    new_head = None
    current = None

    while list1 and list2:
        if list1.data > list2.data:
            new_node = Node(list1.data)
        else:
            new_node = Node(list2.data)

        if new_head is None:
            new_head = new_node
            current = new_node
        else:
            current.next = new_node
            current = new_node

        list1 = list1.next
        list2 = list2.next

    if list1 is not None:
        current.next = list1
    elif list2 is not None:
        current.next = list2

    return new_head

# Example usage
list1 = Node(5)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(8)

list2 = Node(1)
list2.next = Node(7)
list2.next.next = Node(4)
list2.next.next.next = Node(5)

new_head = createNewLinkedList(list1, list2)

# Print the new linked list
temp = new_head
while temp:
    print(temp.data, end=" ")
    temp = temp.next
```

### Answer 2

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def removeDuplicates(head):
    if head is None or head.next is None:
        return head

    current = head
    while current and current.next:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head

# Example usage
# Create the linked list: 11->11->11->21->43->43->60
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

head = removeDuplicates(head)

# Print the modified linked list
temp = head
while temp:
    print(temp.data, end=" ")
    temp = temp.next
```

### Answer 3

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        if head is None or head.next is None or k == 1:
            return head

        # Create a dummy node to serve as the new head of the modified list
        dummy = ListNode(0)
        dummy.next = head
        prev = dummy

        while True:
            # Check if there are at least k nodes remaining
            start = prev.next
            end = prev

            for _ in range(k):
                end = end.next
                if end is None:
                    # Not enough nodes remaining, so return the modified list
                    return dummy.next

            # Reverse the group of k nodes
            future = end.next
            cur = start
            prev = None

            while cur != future:
                next_node = cur.next
                cur.next = prev
                prev = cur
                cur = next_node

            # Update the links
            prev_next = prev.next
            prev.next = future
            prev = prev_next
            prev.next = future

        return dummy.next
```

### Answer 4

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def reverse_sublist(head, k):
    prev = None
    current = head
    next = None

    for _ in range(k):
        next = current.next
        current.next = prev
        prev = current
        current = next

    return prev


def reverse_alternate_k_nodes(head, k):
    if head is None or head.next is None or k <= 1:
        return head

    current = head
    prev = None

    while current:
        count = 0

        # Skip k nodes
        while count < k and current:
            prev = current
            current = current.next
            count += 1

        if count == k:
            # Reverse k nodes
            prev.next = reverse_sublist(prev.next, k)

        count = 0

        # Skip k nodes
        while count < k and current:
            prev = current
            current = current.next
            count += 1

    return head


def print_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()


# Example:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(7)
head.next.next.next.next.next.next.next = ListNode(8)
head.next.next.next.next.next.next.next.next = ListNode(9)

k = 3

print("Input: ", end="")
print_list(head)

head = reverse_alternate_k_nodes(head, k)

print("Output: ", end="")
print_list(head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def delete_last_occurrence(head, key):
    if head is None:
        return head

    prev = None
    last = None
    current = head

    while current:
        if current.val == key:
            last = current
        current = current.next

    if last is None:
        return head

    if last == head:
        head = head.next
    else:
        current = head
        while current.next != last:
            current = current.next
        current.next = last.next

    return head


def print_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()


# Example:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(5)
head.next.next.next.next = ListNode(2)
head.next.next.next.next.next = ListNode(10)

key = 2

print("Input: ", end="")
print_list(head)

head = delete_last_occurrence(head, key)

print("Output: ", end="")
print_list(head)
```

### Answer 6

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def merge_sorted_lists(a, b):
    if a is None:
        return b
    if b is None:
        return a

    if a.val > b.val:
        a, b = b, a

    result = a
    while a.next and b:
        if a.next.val > b.val:
            a.next, b = b, a.next
        a = a.next

    if b:
        a.next = b

    return result


def print_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()


# Example 1:
a = ListNode(5)
a.next = ListNode(10)
a.next.next = ListNode(15)

b = ListNode(2)
b.next = ListNode(3)
b.next.next = ListNode(20)

print("Input:")
print("a: ", end="")
print_list(a)
print("b: ", end="")
print_list(b)

merged = merge_sorted_lists(a, b)

print("Output: ", end="")
print_list(merged)

# Example 2:
a = ListNode(1)
a.next = ListNode(1)

b = ListNode(2)
b.next = ListNode(4)

print("Input:")
print("a: ", end="")
print_list(a)
print("b: ", end="")
print_list(b)

merged = merge_sorted_lists(a, b)

print("Output: ", end="")
print_list(merged)
```

### Answer 7

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


def reverse_doubly_linked_list(head):
    if head is None or head.next is None:
        return head

    current = head
    prev = None

    while current:
        next = current.next
        current.next = prev
        current.prev = next
        prev = current
        current = next

    return prev


def print_list(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()


# Example:
head = Node(10)
head.next = Node(8)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next
head.next.next.next = Node(2)
head.next.next.next.prev = head.next.next

print("Original Linked list: ", end="")
print_list(head)

head = reverse_doubly_linked_list(head)

print("Reversed Linked list: ", end="")
print_list(head)
```

### Answer 8

```python
from collections import Counter

def intersect(nums1, nums2):
    freqMap = Counter(nums1)
    intersection = []

    for num in nums2:
        if freqMap[num] > 0:
            intersection.append(num)
            freqMap[num] -= 1

    return intersection

# Example usage
nums1 = [1, 2, 2, 1]
nums2 = [2, 2]
print(intersect(nums1, nums2))
```
